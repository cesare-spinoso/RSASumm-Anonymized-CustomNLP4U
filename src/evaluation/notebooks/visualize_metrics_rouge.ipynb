```python
for model_name, json_path in yaml_file_contents.items():
    if json_path is not None:
        df = convert_eval_json_to_df(json_path, evals_to_exclude=["weighted_rsa"])
        df = pd.concat([df, df_sota])
        
dataset_dfs = {}
for dataset_name in ["covidet", "debatepedia", "duc_single", "multioped", "qmsum"]:
    df_per_dataset = pd.DataFrame(columns=["random", "best", "direct", "source_reconstruction", "latent_reconstruction"])
    for model_name, json_path in yaml_file_contents.items():
        if json_path is None:
            continue
        df_per_model = convert_eval_json_to_df(json_path, multiple_metrics_format="list", evals_to_exclude=["weighted_rsa"])
        df_per_dataset.loc[len(df_per_dataset)] = df_per_model[dataset_name].values.tolist()
        df_per_dataset.index = df_per_dataset.index.tolist()[:-1] + [model_name]
    df_per_dataset = df_per_dataset.T     
    dataset_dfs[dataset_name] = df_per_dataset 

rouge_metrics = ["rouge1", "rouge2", "rougeL"]
literal_readers = ["random", "direct", "source_reconstruction", "latent_reconstruction"]
aggregate_metrics = {lr : dict.fromkeys(rouge_metrics, 0) for lr in literal_readers}
for dataset_name, df in dataset_dfs.items():
    print("="*80)
    print("Dataset name: ", dataset_name)
    filtered_df = df[["bart_e2e", "led_e2e", "llama2_e2e"]]
    filtered_df = filtered_df[filtered_df.index != "best"]
    for col in filtered_df.columns:
        for i, rm in enumerate(rouge_metrics):
            rouge_column = filtered_df[col].apply(lambda x: x[i])
            row_max = rouge_column.idxmax(axis=0)
            aggregate_metrics[row_max][rm] += 1
df_agg_metrics = pd.DataFrame(aggregate_metrics)
df_agg_metrics = df_agg_metrics.T
pretty_print(df_agg_metrics)
pretty_print(df_agg_metrics/df_agg_metrics.sum(axis=0))
```

```python
for model_name, json_path in yaml_file_contents.items():
    if json_path is not None:
        df = convert_eval_json_to_df(json_path, evals_to_exclude=["weighted_rsa"])
        df = pd.concat([df, df_sota])
        
dataset_dfs = {}
for dataset_name in ["covidet", "debatepedia", "duc_single", "multioped", "qmsum"]:
    df_per_dataset = pd.DataFrame(columns=["random", "best", "direct", "source_reconstruction", "latent_reconstruction"])
    for model_name, json_path in yaml_file_contents.items():
        if json_path is None:
            continue
        df_per_model = convert_eval_json_to_df(json_path, multiple_metrics_format="list", evals_to_exclude=["weighted_rsa"])
        df_per_dataset.loc[len(df_per_dataset)] = df_per_model[dataset_name].values.tolist()
        df_per_dataset.index = df_per_dataset.index.tolist()[:-1] + [model_name]
    df_per_dataset = df_per_dataset.T     
    dataset_dfs[dataset_name] = df_per_dataset 

relative_lat_source = {}
for dataset_name, df in dataset_dfs.items():
    print("="*80)
    print("Dataset name: ", dataset_name)
    relative_lat_source[dataset_name] = {}
    filtered_df = df[["bart_e2e", "led_e2e", "llama2_e2e"]]
    for col in filtered_df.columns:
        for i, rm in enumerate(rouge_metrics):
            rouge_column = filtered_df[col].apply(lambda x: x[i])
            # print(rouge_column)
            ratio = rouge_column["latent_reconstruction"]/rouge_column["source_reconstruction"]
            # print("Ratio ", ratio)
            signed_ratio =  (ratio - 1) * 100
            string_ratio = ("$+$" if signed_ratio >= 0 else "$-$") + f"{abs(signed_ratio):.2f}"
            relative_lat_source[dataset_name][f"{col}-{rm}"] = string_ratio
df_relative_lat_source = pd.DataFrame(relative_lat_source)
df_relative_lat_source = df_relative_lat_source.T
pretty_print(df_relative_lat_source)
```
